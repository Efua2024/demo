```python
df1 = pd.read_csv("Student.csv")
df1
```

```python
df1["total score"] = round(100 * (df1["math score"] + df1["reading score"] + df1["writing score"])/ 300,2)
df1
```

```python
def grade(x):
    if x >= 90 and x <= 100:
        return "A"
    elif x >= 80 and x < 90:
        return "B"
    elif x >= 70 and x < 80:
        return "C"
    elif x >= 60 and x < 70:
        return "D"
    else:
        return "F"
```

```python
df1["grade"] = df1["total score"].apply(grade)
df1
```

```python
#checking number of students with different grades 
sns.countplot(df1, x="grade")
```

```python
#bivariate analysis
plt.figure(figsize=(10,5))
sns.boxplot(df1, x="parental level of education", y="math score")
```
